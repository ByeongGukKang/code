```python
import numpy as np 
import pandas as pd 
```

```python
d = pd.read_pickle("Data/binance_data.pickle")
```

```python
d
```

```python
coin_lst = d.permno.unique().tolist()
coin_lst[:10]
```

```python
busd = [col for col in coin_lst if "BUSD" in col[-4:]]
usdt = [col for col in coin_lst if "USDT" in col[-4:]]
```

```python
len(busd)+len(usdt), len(busd), len(usdt)
```

```python
# 둘다 겹치는 코인
[coin for coin in busd if coin in usdt]
```

```python
# USDT만 있거나 BUSD만 있는 걸 찾아야함
bus = [coin.replace("BUSD","") for coin in busd]
usd = [coin.replace("USDT","") for coin in usdt]
```

```python
find = []
for coin in bus:
    if coin not in usd:
        find.append(coin)
```

```python
find
```

```python
auction = d.loc[d.permno == "AUCTIONBUSD"]
auction.loc[:,"permno"] = auction.loc[:,"permno"].str.replace("BUSD","")

auction
```

```python
# USDT만 가져오자
d2 = d.loc[d.permno.apply(lambda x: x in usdt)]
```

```python
# 둘을 합치자
concat_df = pd.concat([d2, auction]).sort_index()
concat_df
```

```python
# 나중에 알게된 사실인데,, USDC도 드랍해야함
concat_df2 = concat_df.loc[concat_df.permno != "USDC"]
concat_df2
```
